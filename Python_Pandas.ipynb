# Brief History

* The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008. It has function to analyse, clean, explore and manipulate the data.
* It helps to work with large datasets and find the results based on statistical theories.
* Pandas can clean the messy datasets, and make them readable and relevant.
* Pandas can also delete the rows that are not relevant, or containing wrong values, like empty or NULL values. This is called the cleaning the data.

* Pandas is a labeled, columnar, Numpy-backed relational engine with a sophisticated block-oriented memory manager.

* It is a BlockManager that stores heterogeneous data in homogeneous blocks for cache efficiency.
